Дана функция f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30

1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
from sympy import *
init_printing()
```

```python
x = Symbol('x')
expression = -12 * x**4 * sin(cos(x)) - 18 * x**3 + 5 * x**2 + 10 * x - 30
expression
```

```python
print('1. Корни уравнения:\n')
nsolve(expression, 1)
```

```python
print('2. Найти интервалы, на которых функция возрастает\n')
nsolve(expression > 0, 1)
```

```python
print('3. Найти интервалы, на которых функция убывает\n')
nsolve(expression < 0, 1)
```

```python
print('4. Построить график\n')
plot(expression)
expression
```

```python
print('5. Вычислить вершину\n')
plot(diff(expression))
diff(expression)
```

```python
print('6. Определить промежутки, на котором f > 0\n')
nsolve(diff(expression) > 0, 1)
```

```python
print('7. Определить промежутки, на котором f < 0\n')
nsolve(diff(expression) < 0, 1)
```
