Дана функция f(x) = 5x^2 + 10x - 30

3. Найти интервалы, на которых функция убывает

7. Определить промежутки, на котором f < 0

```python
from sympy import *
from sympy.plotting import plot
init_printing()
```

```python
x = Symbol('x')
a = 5
b = 10
c = 30
expression = a * x**2 + b * x - c
plot(expression, show=True)
expression
```

```python
diff(expression)
```

```python
x1, x2 = solve(expression)
```

```python
type(solve(expression))
```

```python
round(x1.evalf(), 3)
```

```python
solve(diff(expression))
```

```python
solve(expression > 0)
```

```python
solve(expression < 0)
```

```python
solve(diff(expression) > 0)
```

```python
solve(diff(expression) < 0)
```
